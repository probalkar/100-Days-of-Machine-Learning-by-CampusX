## Univariate Analysis
Univariate analysis in Exploratory Data Analysis (EDA) focuses on examining individual variables in a dataset independently. It involves studying the distribution, characteristics, and summary statistics of a single variable at a time without considering the relationship with other variables.

### Key Aspects of Univariate Analysis:

#### 1. **Central Tendency:**
   - **Mean, Median, Mode:** Understanding the average or central value around which the data is distributed.
   - **Measures of Dispersion:** Variability around the central value, including range, variance, standard deviation.

#### 2. **Distribution of Data:**
   - **Histograms and Frequency Distributions:** Visualizing the distribution and frequency of values within predefined intervals (bins).
   - **Boxplots:** Displaying the median, quartiles, and outliers within the data.

#### 3. **Summary Statistics:**
   - **Descriptive Statistics:** Summarizing the data with statistics like minimum, maximum, quartiles, etc.
   - **Skewness and Kurtosis:** Assessing asymmetry and shape of the distribution.

#### 4. **Missing Values and Outliers:**
   - **Identifying Missing Values:** Analyzing the presence of missing or null values within the variable.
   - **Outlier Detection:** Identifying extreme values that deviate significantly from the rest of the data.

#### 5. **Visualization Techniques:**
   - **Bar Charts:** Visualizing categorical data.
   - **Histograms and Density Plots:** Displaying the distribution of numerical data.
   - **Boxplots:** Highlighting central tendency, variability, and outliers.

### Purpose of Univariate Analysis in EDA:
- **Initial Exploration:** Understanding each variable's characteristics and nature before deeper analysis.
- **Data Cleaning:** Identifying missing values, outliers, or potential data quality issues.
- **Variable Selection:** Assessing which variables are relevant for further analysis or modeling.

### Example Techniques:
- **For Numerical Variables:** Histograms, boxplots, summary statistics.
- **For Categorical Variables:** Bar charts, frequency tables, mode calculation.

### Importance:
Univariate analysis forms the foundation of exploratory data analysis, providing initial insights into individual variables before delving into multivariate relationships. It helps in understanding the data's characteristics, identifying potential issues, and guiding further analysis or preprocessing steps.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df = df.iloc[:, :8]
df.head()
```

## 1. Categorical

### a. Countplot

```python
# df['Survived'].value_counts().plot(kind='bar')
sns.countplot(x=df['survived'])
```

```python
sns.countplot(x=df['pclass'])
```

```python
sns.countplot(x=df['sex'])
```

```python
sns.countplot(x=df['embarked'])
```

### b. Pie Chart

```python
df['survived'].value_counts().plot(kind='pie', autopct='%.2f%%')
```

```python
df['pclass'].value_counts().plot(kind='pie', autopct='%.2f%%')
```

```python
df['sex'].value_counts().plot(kind='pie', autopct='%.2f%%')
```
