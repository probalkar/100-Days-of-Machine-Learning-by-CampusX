Matplotlib is a popular Python library used for creating static, interactive, and publication-quality visualizations. It offers a wide range of functionalities to visualize data in various formats, including line plots, bar plots, scatter plots, histograms, pie charts, and more.

### Overview of Matplotlib:

1. **Basic Structure:**
   - Matplotlib operates using two main components:
     - **Matplotlib.pyplot:** The scripting layer that provides a MATLAB-like interface for creating plots and visualizations.
     - **Matplotlib Figure and Axes:** The object-oriented interface that allows more control and customization of plots.

2. **Commonly Used Types of Plots:**
   - **Line Plot:** Display data points connected by straight line segments.
   - **Bar Plot:** Represent categorical data using rectangular bars.
   - **Pie Chart:** Display data as a proportion of a whole (wedges of a circle).
   - **Scatter Plot:** Display points or markers at positions specified by x and y coordinates.
   - **Histogram:** Display the distribution of a continuous numerical variable.

### Examples of Creating Different Plots:

#### 1. Pie Chart:
```python
import matplotlib.pyplot as plt

# Data for the pie chart
sizes = [30, 20, 25, 25]
labels = ['A', 'B', 'C', 'D']

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

#### 2. Line Chart with Markers and Marker Size Parameters:
```python
import matplotlib.pyplot as plt

# Data for the line plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 3, 5]

# Create a line plot with markers and marker size
plt.plot(x, y, marker='o', markersize=8)
plt.title('Line Chart with Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)  # Show gridlines
plt.show()
```

#### 3. Bar Graph:
```python
import matplotlib.pyplot as plt

# Data for the bar graph
categories = ['Category A', 'Category B', 'Category C']
values = [20, 35, 30]

# Create a bar graph
plt.bar(categories, values)
plt.title('Bar Graph Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

#### 4. Horizontal Bar Graph:
```python
import matplotlib.pyplot as plt

# Data for the horizontal bar graph
categories = ['Category A', 'Category B', 'Category C']
values = [20, 35, 30]

# Create a horizontal bar graph
plt.barh(categories, values)
plt.title('Horizontal Bar Graph Example')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

These examples demonstrate how to create different types of plots using Matplotlib. Customization options such as labels, titles, axes, markers, and gridlines can be adjusted according to specific visualization requirements. Matplotlib provides extensive documentation and flexibility to create highly customized and informative visualizations for various types of data.

# Pie Chart

````python
areas = ['Marketing', 'Sales', 'Development', 'HR', 'Customer Support']
budget = [2.5, 4, 10, 1, 6]
````

````python
plt.pie(budget, labels=areas, autopct='%0.2f%%')
plt.show()
````

# Scatter Plot

````python
x = [23, 45, 12, 49, 97, 32, 11, 44]
y = [45, 34, 67, 87, 55, 22, 89, 90]

plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary Graph')
````

# Line Chart

````python
product_id = [1,2,3,4,5,6,7,8]
online_price = [233, 456, 770, 120, 222, 444, 200, 300]
offline_price = [400, 300, 100, 333, 444, 565, 565, 899]
chor_bazaar = [100, 200, 300, 400, 150, 250, 450, 500]
````

````python
plt.plot(product_id, online_price, color='red', marker='d', markersize=10, label='Online price')
plt.plot(product_id, offline_price, color='green', marker='+', markersize=10, label='Offline price')
plt.plot(product_id, chor_bazaar, color='blue', marker='.', markersize=10, label='Chor bazaar price')
plt.legend()
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparison')
plt.show()
````

# Bar Graph

````python
year = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
prices = [65000, 69000, 75000, 55000, 52000, 64000, 81000]
plt.bar(year, prices)
plt.xlabel('Year')
plt.ylabel('Price')
````
