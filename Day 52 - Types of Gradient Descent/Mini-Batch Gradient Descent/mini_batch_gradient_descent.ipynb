## Mini-Batch Gradient Descent

Mini-batch gradient descent is a variation of gradient descent where the model parameters (coefficients and intercept) are updated using gradients computed from a small subset of the training dataset, known as mini-batches. This approach combines the efficiency of stochastic gradient descent (SGD) with the stability of batch gradient descent. Mini-batch gradient descent is widely used in practice for training deep learning models and other large-scale machine learning tasks.

### Interpretation:

- Mini-batch gradient descent combines the efficiency of stochastic gradient descent (SGD) with the stability of batch gradient descent.
- It allows for parallelization and efficient computation, making it suitable for training large-scale models on modern hardware.
- The size of the mini-batch is a hyperparameter that needs to be tuned based on the dataset size, computational resources, and convergence behavior.

### Summary:

Mini-batch gradient descent is an optimization algorithm used to minimize the cost function for linear regression with multiple features. By updating the parameters using gradients computed from mini-batches of the training dataset, mini-batch gradient descent strikes a balance between the computational efficiency of stochastic gradient descent and the stability of batch gradient descent.

```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X, y = load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=2)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
r2_score(y_test, y_pred)
```

```python
import random

class MBGDRegressor:
    def __init__(self, batch_size, learning_rate = 0.01, epochs = 100):
        self.coef_ = None
        self.interecept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        self.bs = batch_size

    def fit(self, X_train, y_train):
        # init your coefs
        self.interecept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        batches = int(X_train.shape[0]/self.bs)

        for i in range(self.epochs):
            for j in range(batches):
                #update all the coef and intercept
                idx = random.sample(range(X_train.shape[0]), self.bs)
                y_hat = np.dot(X_train[idx], self.coef_) + self.interecept_
                # print("Shape of y_hat: ", y_hat.shape)
                interecept_der = -2 * np.mean(y_train[idx] - y_hat)
                self.interecept_ = self.interecept_ - (self.lr * interecept_der)
    
                coef_der = -2 * np.dot((y_train[idx] - y_hat), X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)

        print(self.coef_, self.interecept_)

    def predict(self, X_test):
        return np.dot(X_test, self.coef_) + self.interecept_
```

```python
mbgdr = MBGDRegressor(int(X_train.shape[0]/10) ,0.1, 50)
```

```python
mbgdr.fit(X_train, y_train)
```

```python
y_pred = mbgdr.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
