## Categorical Scatterplots
Categorical scatter plots in Seaborn, particularly strip plots and swarm plots, are used to visualize the relationship between one categorical variable and one continuous or numeric variable. They display individual data points for each category, aiding in understanding the distribution and density of the data within each category.

Seaborn provides several functions for creating categorical scatterplots, including `catplot()`, `stripplot()`, and `swarmplot()`.

### `catplot()`:
The `catplot()` function in Seaborn is a powerful tool that serves as a higher-level interface to create various categorical plots by specifying the kind of plot as an argument. It allows you to create different types of categorical plots based on the provided data and chosen plot type.

#### `catplot()` Function Syntax:
```python
sns.catplot(x=None, y=None, hue=None, data=None, kind='strip', ...)
```
- `x`, `y`: Variables for the x-axis and y-axis, respectively.
- `hue`: Optional parameter to add a categorical variable that colors the points based on different categories.
- `data`: DataFrame containing the data.
- `kind`: Type of plot to create (`'strip'` by default for stripplot, or `'swarm'` for swarmplot).

### Strip Plot(`sns.stripplot()`):
A strip plot displays individual data points along a categorical axis. It represents each data point with a dot, jittering the points along the categorical axis to avoid overlapping.

#### `stripplot()` Function Syntax:
```python
sns.stripplot(x=None, y=None, hue=None, data=None, ...)
```
- `x`, `y`: Variables for the x-axis and y-axis, respectively.
- `hue`: Optional parameter to add a categorical variable that colors the points based on different categories.
- `data`: DataFrame containing the data.

### Swarm Plot(`sns.swarmplot()`):
A swarm plot is similar to a strip plot but arranges the points more evenly, ensuring that they do not overlap. It positions the points such that they do not overlap with other points in the same category.

#### `swarmplot()` Function Syntax:
```python
sns.swarmplot(x=None, y=None, hue=None, data=None, ...)
```
- `x`, `y`: Variables for the x-axis and y-axis, respectively.
- `hue`: Optional parameter to add a categorical variable that colors the points based on different categories.
- `data`: DataFrame containing the data.

### Additional Parameters:
- These functions have additional parameters to control various aspects of the plot, such as the color palette, marker styles, ordering of categories, jitter, size, and more.

### Key Differences:
- `stripplot()` creates a scatterplot where data points are stacked along the categorical axis with optional jitter to prevent overlap.
- `swarmplot()` also creates a categorical scatterplot but avoids overlapping points by adjusting their positions based on the distribution of data.

Both `stripplot()` and `swarmplot()` are useful for visualizing categorical data and distributions of observations within categories, with `swarmplot()` often providing a clearer representation of individual data points when dealing with dense data.

The `catplot()` function, on the other hand, serves as a higher-level interface that allows you to create various types of categorical plots by specifying the `kind` parameter. It can be used to generate strip plots (`kind='strip'`) or swarm plots (`kind='swarm'`) and other categorical plot types depending on the specified `kind` argument.

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
plt.style.use('fivethirtyeight')
````

````python
tips = sns.load_dataset('tips')
````

````python
tips.head()
````

By default `kind` parameter of `catplot()` is set to `'strip'`

````python
sns.catplot(x='day', y='tip', kind='strip', data=tips)
````

````python
sns.catplot(x='day', y='tip', kind='strip', jitter=0, data=tips)
````

````python
sns.stripplot(x='day', y='tip', jitter=0.1, data=tips)
````

````python
sns.catplot(x='day', y='tip', kind='swarm', hue='sex', data=tips)
````

````python
sns.swarmplot(x='day', y='tip', data=tips)
````
