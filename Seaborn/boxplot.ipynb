## Boxplots
Boxplots, also known as box-and-whisker plots, are statistical diagrams that display the distribution of numerical data based on five key summary statistics: minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. They provide a concise summary of the data's central tendency, variability, and identify potential outliers.

<img src='https://miro.medium.com/max/9000/1*2c21SkzJMf3frPXPAR_gZA.png'/>

### Anatomy of a Boxplot:
- **Box:** Represents the interquartile range (IQR) between the first quartile (Q1) and third quartile (Q3). It spans the central 50% of the data. The median (Q2) is marked by a line within the box.
- **Whiskers:** Extend from the box to the minimum and maximum values within a specified range (typically 1.5 times the IQR).
- **Outliers:** Individual data points lying beyond the whiskers are considered outliers and are displayed as individual points or dots.

### Use Cases and Interpretation:
- **Central Tendency:** The median line within the box represents the middle value of the dataset.
- **Variability:** The length of the box (IQR) indicates the variability of the data.
- **Skewness and Outliers:** Whiskers help identify the data's spread and presence of outliers.
- **Comparison:** Boxplots are useful for comparing distributions of different groups or categories.

### Creating Boxplots in Seaborn (`sns.boxplot()`):
Seaborn provides a simple and effective way to create boxplots to visualize the distribution of numerical data across different categories.

#### Syntax:
```python
sns.boxplot(x=None, y=None, hue=None, data=None, ...)
```

#### Key Parameters:
- `x`, `y`: Variables defining the categorical and numeric axes, respectively.
- `hue`: An optional parameter that adds a categorical differentiation (adding colors) within the plot.
- `data`: DataFrame containing the data.
- Additional parameters for customization such as `palette`, `width`, `showfliers`, etc.

Boxplots provide a concise summary of the distribution of numerical data, making them valuable tools for visualizing and comparing distributions across different categories or groups, identifying outliers, and understanding the spread and central tendency of the data. Adjust parameters for better insights and data exploration.

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
plt.style.use('fivethirtyeight')
````

````python
tips = sns.load_dataset('tips')
````

````python
tips.head()
````

````python
# Univariate
sns.boxplot(tips['tip'])
````

````python
sns.boxplot(tips['total_bill'])
````

````python
# Multivariate
sns.catplot(x='day', y='total_bill', kind='box', data=tips)
````

````python
sns.catplot(x='day', y='total_bill', kind='box', hue='sex', data=tips)
````
