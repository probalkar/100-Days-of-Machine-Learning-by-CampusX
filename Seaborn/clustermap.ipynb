## Clustermap
Clustermaps in Seaborn are specialized heatmaps that perform hierarchical clustering to rearrange rows and columns in a dataset based on their similarity. They visualize a matrix dataset by clustering similar rows and columns together, allowing patterns and relationships within the data to become more apparent.

### Key Features of Clustermaps:
- **Hierarchical Clustering:** Utilizes hierarchical clustering algorithms to reorder rows and columns based on similarity.
- **Color Representation:** Represents values in the dataset with a color gradient, similar to regular heatmaps.
- **Dendrograms:** Displays dendrograms along rows and columns showing the hierarchical relationships.

### Parameters Used in Seaborn's `clustermap()` Function:

#### `data` (Data):
- **Purpose:** The input data for the clustermap.

#### `cmap` (Colormap):
- **Purpose:** Specifies the color palette or colormap to be used in the clustermap.
- **Values:** Various colormaps are available (`'viridis'`, `'coolwarm'`, `'YlGnBu'`, etc.).

#### `standard_scale`:
- **Purpose:** Standardizes the data along rows or columns before clustering.
- **Values:** `0` or `'row'` normalizes each row, `1` or `'column'` normalizes each column.

#### `method` (Clustering Method):
- **Purpose:** Determines the hierarchical clustering algorithm to use.
- **Values:** `'single'`, `'complete'`, `'average'`, `'ward'`, etc.

#### `cbar` (Color Bar):
- **Purpose:** Determines whether to display a color bar alongside the clustermap.
- **Values:** `True` displays the color bar, `False` hides it.

#### `annot` (Annotations):
- **Purpose:** Controls the display of numerical annotations within the cells of the clustermap.
- **Values:** `True` displays annotations, `False` hides them.

### Usefulness:
- **Pattern Identification:** Reveals patterns and relationships within complex datasets.
- **Data Organization:** Hierarchically clusters similar rows and columns together.
- **Visualization of Similarity:** Demonstrates similarity or dissimilarity between rows and columns.

Clustermaps are powerful tools for exploring and understanding complex datasets, revealing inherent patterns and relationships by rearranging rows and columns based on similarity or dissimilarity. Adjust the parameters to effectively visualize and interpret the hierarchical clustering within your dataset.

```python
import seaborn as sns
import pandas as pd
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x = flights.pivot(index='year', columns='month', values='passengers')
```

```python
x
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, annot=True, metric='correlation', z_score=0, row_cluster=False)
```
