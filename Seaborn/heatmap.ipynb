## Heatmaps
Heatmaps in Seaborn are graphical representations of data where values in a matrix are represented as colors. They are particularly useful for visualizing relationships and patterns in data, especially when dealing with large datasets. Heatmaps use color gradients to represent the magnitude of values, making it easier to identify high or low values and patterns within the data.

### Key Features of Heatmaps:
- **Color Representation:** Each cell in the matrix is represented by a color, with the color intensity or shade indicating the value.
- **Two-Dimensional Representation:** Rows and columns represent variables, and cell colors represent the relationship between these variables.
- **Visualizing Patterns:** Heatmaps help identify correlations, clusters, or patterns in the data.

### Parameters Used in Seaborn's `heatmap()` Function:

#### `cbar` (Color Bar):
- **Purpose:** Determines whether to display a color bar alongside the heatmap to indicate the scale of colors.
- **Values:** `True` displays the color bar, `False` hides it.

#### `cmap` (Colormap):
- **Purpose:** Specifies the color palette or colormap to be used in the heatmap.
- **Values:** Various colormaps are available (`'viridis'`, `'coolwarm'`, `'YlGnBu'`, etc.).

#### `fmt` (Format String for Annotations):
- **Purpose:** Determines the format of the annotations in the cells of the heatmap.
- **Values:** String formatting options for cell annotations (e.g., `'.2f'` for two decimal places).

#### `annot` (Annotations):
- **Purpose:** Controls the display of numerical annotations within the cells of the heatmap.
- **Values:** `True` displays annotations, `False` hides them.

#### `linewidths` (Line Widths):
- **Purpose:** Specifies the width of the lines that will divide each cell in the heatmap.
- **Values:** Numeric value defining the width of the lines.

### Usefulness:
- **Correlation Analysis:** Visualizing correlations between variables.
- **Pattern Recognition:** Identifying clusters or trends in data.
- **Comparative Analysis:** Comparing values across two dimensions.

Seaborn's `heatmap()` function provides numerous parameters for customizing the appearance and information displayed within the heatmap, offering a powerful tool for visualizing data relationships and patterns in a matrix format. Adjust these parameters to effectively visualize and interpret the data based on specific analysis requirements.

```python
import seaborn as sns
import pandas as pd
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='viridis')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='summer')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='winter')
```
