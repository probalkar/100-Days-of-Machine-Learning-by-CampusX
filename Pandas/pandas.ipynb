# Pandas

Pandas is a powerful open-source data manipulation and analysis library built on top of Python. It provides easy-to-use data structures and tools for working with structured data, primarily in the form of two key data structures: Series and DataFrame.

**Properties:**

1. **DataFrame:** The central data structure in Pandas is the DataFrame, a 2-dimensional labeled data structure with columns that can hold different types of data. It resembles a spreadsheet or SQL table and allows easy manipulation and analysis of data.

2. **Series:** Series is a one-dimensional labeled array capable of holding data of any type. It's often used to represent columns in a DataFrame or single-dimensional data structures.

3. **Data Alignment:** Pandas handles data alignment automatically along both the rows and columns when performing operations on DataFrames and Series, aligning data based on label indices.

4. **Missing Data Handling:** Pandas provides robust methods to handle missing or NaN (Not a Number) values in datasets, allowing for flexible data cleaning and manipulation.

5. **Data I/O:** It supports reading and writing data from various file formats, such as CSV, Excel, SQL databases, and more, making it easy to work with data from different sources.

**Applications:**

1. **Data Cleaning and Preparation:** Pandas is extensively used in data preprocessing tasks like cleaning messy data, handling missing values, transforming data, and preparing it for analysis.

2. **Data Analysis and Exploration:** It's widely used for data exploration, descriptive statistics, aggregations, filtering, grouping, and other data analysis tasks due to its powerful indexing and manipulation capabilities.

3. **Time Series Analysis:** Pandas offers specialized tools for working with time series data, making it popular for tasks like financial analysis, stock market analysis, and forecasting.

4. **Data Visualization:** While Pandas itself doesn’t provide visualization capabilities, it integrates seamlessly with visualization libraries like Matplotlib and Seaborn, allowing users to create informative plots and charts from Pandas data structures.

5. **Machine Learning and Statistical Analysis:** It's used in conjunction with machine learning libraries like Scikit-learn and for statistical analysis due to its ability to handle structured data efficiently.

In summary, Pandas is a versatile and essential tool for data manipulation, cleaning, exploration, and analysis in data science, research, finance, business analytics, and many other domains where structured data handling is crucial.

<h3>The .isin() function</h3>
Find the most destructive death over batsman in the history of IPL

Strike rate = (Number of runs/Number of balls)*100
<br>
Minimum balls played should be 200 and between over 16-20

````python
death_over = delivery[delivery['over']>15]
````

````python
all_batsmen = death_over.groupby('batsman')['batsman_runs'].count()
x = all_batsmen>200
batsman_list = all_batsmen[x].index.to_list()
````

````python
final = delivery[delivery['batsman'].isin(batsman_list)]
````

````python
# Runs scored by all these 43 batsmen
runs = final.groupby('batsman')['batsman_runs'].sum()
````

````python
# Balls played by all these 43 batsmen
balls = final.groupby('batsman')['batsman_runs'].count()
````

````python
# Calculating strike rate
sr = (runs/balls)*100
sr.sort_values(ascending=False).index[0]
````

<h3>dropna() Method</h3>
Used to drop missing values from the dataset

**Description:**
The `dropna()` function is a method available on Pandas DataFrames that allows you to eliminate rows or columns that contain missing values (NaN).

**Parameters:**

- `axis` (default: `0`): Specifies whether to drop rows (`0` or `'index'`) or columns (`1` or `'columns'`) that contain null values.
- `how` (default: `'any'`): Specifies the condition for dropping. Options include:
    - `'any'`: Drops the row or column if it contains any NaN values.
    - `'all'`: Drops the row or column only if all values are NaN.
- `thresh` (default: `None`): Specifies the minimum number of non-null values required to keep a row or column. If the number of non-null values falls below this threshold, the row or column is dropped.
- `subset` (default: `None`): Specifies a subset of columns or rows to consider for null value dropping.
- `inplace` (default: `False`): Modifies the DataFrame in place if set to `True`.

**Usage:**
```python
# Drop rows with NaN values
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

# Drop columns with NaN values
df.dropna(axis=1, how='any', thresh=None, subset=None, inplace=False)
```

**Example:**
Consider a DataFrame `df`:
```python
import pandas as pd

data = {'A': [1, 2, None, 4],
        'B': [5, None, 7, 8],
        'C': [9, 10, 11, 12]}

df = pd.DataFrame(data)

# Drop rows with any NaN values
cleaned_df = df.dropna()

# Drop columns with any NaN values
cleaned_df_cols = df.dropna(axis=1)
```

In this example:
- `cleaned_df` will contain rows where all columns have non-null values.
- `cleaned_df_cols` will contain columns where all rows have non-null values.

The `dropna()` function is valuable for data cleaning, especially when dealing with datasets containing missing values.

<h3>fillna() Method</h3>

Used to fill the missing values in the dataset

**Description:**
The `fillna()` method in Pandas allows you to replace NaN (missing) values in a DataFrame or Series with specific values.

**Parameters:**

- `value`: Specifies the value to use for replacing NaN occurrences.
- `method` (default: `None`): Specifies a method to use for filling missing values. Options include:
    - `'ffill'` or `'pad'`: Forward fills the missing values with the last known value.
    - `'bfill'` or `'backfill'`: Backward fills the missing values with the next known value.
- `axis` (default: `0`): Specifies the axis along which to fill missing values (`0` for rows, `1` for columns).
- `inplace` (default: `False`): Modifies the DataFrame in place if set to `True`.
- `limit` (default: `None`): Specifies the maximum number of consecutive NaN values to fill when using the `method` parameter.

**Usage:**
```python
# Fill NaN values with a specific value
df.fillna(value=None, method=None, axis=None, inplace=False, limit=None)
```

**Example:**
Consider a DataFrame `df`:
```python
import pandas as pd
import numpy as np

data = {'A': [1, 2, None, 4],
        'B': [5, None, 7, 8],
        'C': [9, 10, 11, None]}

df = pd.DataFrame(data)

# Fill NaN values with a specific value (e.g., 0)
filled_df = df.fillna(0)

# Fill NaN values using forward fill (ffill)
ffilled_df = df.fillna(method='ffill')

# Fill NaN values using backward fill (bfill)
bfilled_df = df.fillna(method='bfill')
```

In this example:
- `filled_df` will contain the DataFrame with NaN values replaced by zeros.
- `ffilled_df` will have NaN values replaced by the last known values in the respective columns (forward fill).
- `bfilled_df` will have NaN values replaced by the next known values in the respective columns (backward fill).

The `fillna()` function is useful for handling missing data by either replacing NaN values with a specific constant or filling them using methods like forward or backward filling based on existing values in the DataFrame.
